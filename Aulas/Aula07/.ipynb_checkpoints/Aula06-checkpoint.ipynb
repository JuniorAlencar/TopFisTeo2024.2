```python
import random
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### Deposição sem rejeição em 1D

```python
N = 10
L = 20.0
sigma = 0.75
n_runs = 1000
data = []
```

```python
for run in range(n_runs):
    y = [random.uniform(0.0, L - 2 * N * sigma) for k in range(N)]
    y.sort()
    data += [y[i] + (2 * i + 1) * sigma for i in range(N)]
```

```python
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$rho(x)$', fontsize=14)
plt.title('Densidade de %i discos ($\sigma$=%s) em uma linha de comprimento L=%s' % (N, sigma, L))
plt.hist(data, bins=200, density=True)
plt.savefig('./img/plot-direct_disks_noreject.png')
```

### Resultado analítico

```python
def binomialCoeff(n, k):
    result = 1
    for i in range(1, k+1):
        result = result * (n-i+1) / i
    return result
```

```python
def Z(N, L, sigma):
    freespace = L - 2.0 * N * sigma
    if freespace > 0.0:
        result = freespace ** N
    else:
        result = 0.0
    return result
```

```python
def pi(x, N, L, sigma):
    tot = 0.
    for k in range(0, N):
        Z1 = Z(k, x - sigma, sigma)
        Z2 = Z(N - k - 1, L - x - sigma, sigma)
        tot += binomialCoeff( N - 1, k) * Z1 * Z2
    Ztotal = Z(N, L, sigma)
    return tot / Ztotal
```

```python
L = 20.0
N = 10
sigma = 0.75
xr = np.linspace(0.0, L, 201)
yr = [pi(x, N, L, sigma) for x in xr]
```

```python
plt.plot(xr, yr, 'red', linewidth=2.0)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$rho(x)$', fontsize=14)
plt.title('Densidade de %i discos ($\sigma$=%s) em uma linha de comprimento L=%s' % (N, sigma, L))
plt.hist(data, bins=200, density=True)
plt.savefig('./img/plot-comparison_disks_noreject.png')
```
